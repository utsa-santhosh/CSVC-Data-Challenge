```python
# function that makes you able to get the names of all students in a class
def make_arr_kids (className): 
    df_cohort.Cohort ==className 
    class_df = df_cohort[df_cohort.Cohort == className]
    kids = class_df['Name'].to_numpy()
    return kids
```

```python
# function that returns the mean value RIT score for a class
def return_mean_score(season, names):
    df_kids = season[season['Name'].isin(names)]
    return df_kids['RITScore'].mean()
```

```python
def get_seasonal_average(season, teachers):
    kids = make_arr_kids(teachers)
    df_small = season[season['Name'].isin(kids)]
    return df_small
```

```python
def percent_met_growth(df):
    sum = 0
    df['FallToWinterMetProjectedGrowth']
    df_clean = df[df['FallToWinterMetProjectedGrowth']=='Yes*']
    sum = len(df_clean.index)
    df_clean2 = cam_winter_math[cam_winter_math['FallToWinterMetProjectedGrowth']=='Yes']
    sum = sum+ len(df_clean2.index)
    total = len(df.index)
    return (sum/total)*100
```

```python
def get_num_low(df, str):
    low = 0
    low_avg = 0
    df_low = df[df[str]=='Low']
    df_low_avg = df[df[str]=='LoAvg']
    low = len(df_low.index)
    low_avg = len(df_low_avg.index)
    return low+low_avg
```

```python
def get_lowest_obj_math (df):
    Operations_and_Algebraic_Thinking = get_num_low(df, 'Goal1Adjective')
    The_Real_and_Complex_Number_Systems = get_num_low(df, 'Goal2Adjective')
    Geometry = get_num_low(df, 'Goal3Adjective')
    Statistics_and_Probability = get_num_low(df, 'Goal4Adjective')
    var = {Operations_and_Algebraic_Thinking: 'Operations and Algebraic Thinking', 
           The_Real_and_Complex_Number_Systems: 'The Real and Complex Number Systems',
           Geometry: 'Geometry',
           Statistics_and_Probability: 'Statistics and Probability'}
    return var.get(max(var))
```

```python
def get_lowest_obj_ela (df):
    Key_Ideas_and_Details = get_num_low(df, 'Goal1Adjective')
    Language_Craft_and_Structure = get_num_low(df, 'Goal2Adjective')
    info_Key_Ideas_and_Details = get_num_low(df, 'Goal3Adjective')
    info_Language_Craft_and_Structure = get_num_low(df, 'Goal4Adjective')
    vocab = get_num_low(df, 'Goal5Adjective')
    var = {Key_Ideas_and_Details: 'Literary Text: Key Ideas and Details', 
           Language_Craft_and_Structure: 'Literary Text: Language, Craft, and Structure',
           info_Key_Ideas_and_Details: 'Informational Text: Key Ideas and Details',
           info_Language_Craft_and_Structure: 'Informational Text: Language, Craft, and Structure',
          vocab: 'Vocabulary: Acquisition and Use'}
    return var.get(max(var))
```
